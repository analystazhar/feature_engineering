```python
import numpy as np
import pandas as pd
```

```python
df1=pd.read_csv('Supermart Grocery Sales - Retail Analytics Dataset.csv')
```

```python
df1.head()
```

```python
new_df=df1[['Order Date']]
```

```python
new_df.head()
```

```python
new_df.info()
```

# by default the data type of the order date is string, so we have to convert this into the datetype 

```python
new_df['Order Date']=pd.to_datetime(new_df['Order Date'])
```

```python
new_df.info()
```

# 1. extract year

```python
 new_df['date_year'] = new_df['Order Date'].dt.year
```

```python
new_df
```

# 2. extract months number

```python
new_df['date_months_number'] = new_df['Order Date'].dt.month
```

```python
new_df
```

# 3. extracting month name

```python
new_df['month_name'] = new_df['Order Date'].dt.month_name()
```

```python
new_df
```

# extract days

```python
new_df['date_day'] = new_df['Order Date'].dt.day
```

```python
new_df
```

```python
#day of week
new_df['day_of_the_week'] = new_df['Order Date'].dt.dayofweek
```

```python
new_df
```

```python
#day of week name
new_df['dow_name'] = new_df['Order Date'].dt.day_name()
```

```python
new_df
```

```python
#is weekend ?
new_df['date_is_weekend'] = np.where(new_df['dow_name'].isin(['Sunday','Saturday']),1,0)
```

```python
new_df
```

```python
# extract of the year
```

```python
new_df['date_week'] = new_df['Order Date'].dt.week
```

```python
new_df
```

```python
new_df['quarter'] = new_df['Order Date'].dt.quarter 
```

```python
new_df
```

```python
new_df['semester'] = np.where(new_df['quarter'].isin([1,2]),1,2)
```

```python
new_df
```
